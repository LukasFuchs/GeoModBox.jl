$\begin{equation}
\rho c_p\frac{\partial T}{\partial t} = -\frac{\partial q_x}{\partial x}-\frac{\partial q_y}{\partial y} + Q
\end{equation}$ 

$\begin{equation}
\rho c_p\frac{\partial T}{\partial t} + \frac{\partial q_x}{\partial x}+ \frac{\partial q_y}{\partial y} - Q = r
\end{equation}$ 

# Crank-Nicholson - General solution

$\begin{equation}
\frac{\partial{T}}{\partial{t}}=\frac{\kappa}{2}\left(\frac{\partial{\left(T^{n+1} + T^{n}\right)^2}}{\partial^2{x}} + \frac{\partial{\left(T^{n+1} + T^{n}\right)^2}}{\partial^2{y}}\right) + \frac{Q}{\rho c_p}
\end{equation}$

$\begin{equation}
\frac{\partial{T}}{\partial{t}}=\frac{\kappa}{2}\left(\frac{\partial{T^{n+1}}^2}{\partial^2{x}}+\frac{\partial{T^{n}}^2}{\partial^2{x}}+\frac{\partial{T^{n+1}}^2}{\partial^2{y}}+\frac{\partial{T^{n}}^2}{\partial^2{y}}\right) + \frac{Q}{\rho c_p}
\end{equation}$

$\begin{equation}
\frac{\partial{T}}{\partial{t}}=\frac{\kappa}{2}\left(\frac{\partial{T^{n+1}}^2}{\partial^2{x}}+\frac{\partial{T^{n+1}}^2}{\partial^2{y}}\right) + \frac{\kappa}{2}\left(\frac{\partial{T^{n}}^2}{\partial^2{x}}+\frac{\partial{T^{n}}^2}{\partial^2{y}}\right) + \frac{Q}{\rho c_p}
\end{equation}$

$\begin{equation}
\rho c_p\frac{\partial{T}}{\partial{t}}=\frac{k}{2}\left(\frac{\partial{T^{n+1}}^2}{\partial^2{x}}+\frac{\partial{T^{n+1}}^2}{\partial^2{y}}\right) + \frac{k}{2}\left(\frac{\partial{T^{n}}^2}{\partial^2{x}}+\frac{\partial{T^{n}}^2}{\partial^2{y}}\right) + Q
\end{equation}$

$\begin{equation}
\rho c_p\frac{\partial{T}}{\partial{t}} = 
\frac{1}{2}\left(
    \frac{\partial}{\partial{x}}\left(
        k\frac{\partial{T^{n+1}}}{\partial{x}}\right) 
    +\frac{\partial}{\partial{y}}\left(
        k\frac{\partial{T^{n+1}}}{\partial{y}}\right)
    \right) 
+\frac{1}{2}\left(
    \frac{\partial}{\partial{x}}\left(
        k\frac{\partial{T^{n}}}{\partial{x}}\right) 
    +\frac{\partial}{\partial{y}}\left(
        k\frac{\partial{T^{n}}}{\partial{y}}\right)
    \right) + Q
\end{equation}$

$\begin{equation}
\rho c_p\frac{\partial{T}}{\partial{t}} = 
\frac{1}{2}\left(
    -\frac{\partial{q_x^{n+1}}}{\partial{x}} 
    -\frac{\partial{q_y^{n+1}}}{\partial{y}}
\right) 
+\frac{1}{2}\left(
    -\frac{\partial{q_x^{n}}}{\partial{x}}
    -\frac{\partial{q_y^{n}}}{\partial{y}}
\right) + Q
\end{equation}$

$\begin{equation}
\rho c_p\frac{\partial{T}}{\partial{t}} = 
-\frac{1}{2}\frac{\partial{q_i^{n+1}}}{\partial{x_i}} 
-\frac{1}{2}\frac{\partial{q_i^{n}}}{\partial{x_i}}
+Q
\end{equation}$

$\begin{equation}
\rho c_p\frac{\partial{T}}{\partial{t}} = 
-\left(1-C\right)\frac{\partial{q_i^{n+1}}}{\partial{x_i}} 
-C\frac{\partial{q_i^{n}}}{\partial{x_i}}
+Q
\end{equation}$

$\begin{equation}
\rho c_p\frac{\partial{T}}{\partial{t}}  
+\left(1-C\right)\frac{\partial{q_i^{n+1}}}{\partial{x_i}} 
+C\frac{\partial{q_i^{n}}}{\partial{x_i}}
-Q
=r
\end{equation}$

$\begin{equation}
\mathbf{K_1}T^{n+1}=\mathbf{K_2}T^{n}+\frac{Q}{\rho c_p}
\end{equation}$

$\begin{equation}
\mathbf{K_1}T^{n+1}-\mathbf{K_2}T^{n}-\frac{Q}{\rho c_p}=r
\end{equation}$

# Variable thermal parameters

## General Solver 

$\begin{equation}
\rho c_p\frac{\partial{T}}{\partial{t}}  
+\left(1-C\right)\frac{\partial{q_i^{n+1}}}{\partial{x_i}} 
+C\frac{\partial{q_i^{n}}}{\partial{x_i}}
-Q = r
\end{equation}$

$\begin{equation}\begin{gather*}
& \rho_{i,j} c_{p,(i,j)}\frac{\partial{T}}{\partial{t}}\vert{}_{i,j}  
+\left(1-C\right)\left(
    \frac{\partial{q_x^{n+1}}}{\partial{x}}\vert{}_{i,j} +
    \frac{\partial{q_y^{n+1}}}{\partial{y}}\vert{}_{i,j} \right) \\ &
+C\left(
    \frac{\partial{q_x^{n}}}{\partial{x}}\vert{}_{i,j} +
    \frac{\partial{q_y^{n}}}{\partial{y}}\vert{}_{i,j} \right)
-Q_{i,j} = r_{ii}
\end{gather*}\end{equation}$

$\begin{equation}
q_i=-k_i\frac{\partial{T}}{\partial{x_i}}
\end{equation}$

$\begin{equation}
\frac{\partial{q_i}}{\partial{x_i}}=\frac{\partial}{\partial{x_i}}\left(-k_i\frac{\partial{T}}{\partial{x_i}}\right)
\end{equation}$

$\begin{equation}
\frac{\partial{q_x}}{\partial{x}}=
\frac{
    -k_{x,(i+1,j)}\frac{\partial{T}}{\partial{x}}\vert_{i+1,j}
    +k_{x,(i,j)}\frac{\partial{T}}{\partial{x}}\vert_{i,j}}
{\Delta{x}}
\end{equation}$

$\begin{equation}
\frac{\partial{q_y}}{\partial{y}}=
\frac{
    -k_{y,(i,j+1)}\frac{\partial{T}}{\partial{y}}\vert_{i,j+1}
    +k_{y,(i,j)}\frac{\partial{T}}{\partial{y}}\vert_{i,j}}
{\Delta{y}}
\end{equation}$

$\begin{equation}\begin{gather*}
& \rho_{i,j} c_{p,(i,j)}\frac{T_{i,j}^{n+1}-T_{i,j}^{n}}{\Delta{t}} \\ & 
+\left(1-C\right)\left(
    \frac{
        q_{x,(i+1,j)}^{n+1} -
        q_{x,(i,j)}^{n+1}}
    {\Delta{x}} +
    \frac{
        q_{y,(i,j+1)}^{n+1}-
        q_{y,(i,j)}^{n+1}}
    {\Delta{y}} \right) \\ &
+C\left(
    \frac{
        q_{x,(i+1,j)}^{n} -
        q_{x,(i,j)}^{n}}
    {\Delta{x}} +
    \frac{
        q_{y,(i,j+1)}^{n}-
        q_{y,(i,j)}^{n}}
    {\Delta{y}} \right) \\ &
-Q_{i,j} = r
\end{gather*}\end{equation}$
