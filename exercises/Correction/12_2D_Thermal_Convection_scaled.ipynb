# Thermische Konvektion ‚Äì skaliert

Ein Vorteil numerischer und analoger Modelle ist ihre Skalierbarkeit, d.‚ÄØh. ihre **Nicht-Dimensionalisierung** (*Non-Dimensionalisation*). Durch die Einf√ºhrung geeigneter Skalierungskonstanten lassen sich alle physikalischen Einheiten aus den Gleichungen entfernen. Je nach getroffener Annahme bleiben dabei bestimmte dimensionslose Parameter erhalten, durch die sich das physikalische Verhalten gut charakterisieren l√§sst.

## Problemstellung

Wir betrachten erneut eine thermische Konvektion, wie in der [vorherigen √úbung](./11_2D_Thermal_Convection.ipynb).

Unter den gleichen Annahmen (z.‚ÄØB. *Boussinesq*-N√§herung) ergeben sich f√ºr den zweidimensionalen Fall folgende Gleichungen:

**Temperaturerhaltung:**

$\begin{equation}
\rho_0 c_p \left(\frac{\partial T}{\partial t} + v_j \frac{\partial{T}}{\partial{x_j}}\right) = -\frac{\partial q_i}{\partial x_i} + \rho_0 H,
\end{equation}$

wobei:
- $\rho_0$ die Referenzdichte [kg/m¬≥],
- $c_p$ die spezifische W√§rmekapazit√§t [J/kg/K],
- $T$ die Temperatur [K],
- $t$ die Zeit [s],
- $v_j$ die Geschwindigkeit in Raumrichtung $j$ [m/s],
- $q_i$ der W√§rmefluss in Richtung $i$ [W/m¬≤],
- $H$ die W√§rmeproduktionsrate pro Masse [W/kg] ist.

**Impulserhaltung:**

$\begin{equation}
0 = -\frac{\partial{P_{\text{dyn}}}}{\partial{x_{i}}} + \frac{\partial{\tau_{ij}}}{\partial{x_j}} - \rho_0 g_{i} \alpha T,
\end{equation}$

wobei:
- $P_{\text{dyn}}$ der dynamische Druck [Pa],
- $\tau_{ij}$ der Cauchy‚Äôsche Spannungstensor [Pa],
- $g_i$ die Komponente des Schwerebeschleunigungsvektors [m/s¬≤],
- $\alpha$ der thermische Ausdehnungskoeffizient [1/K] ist.

**Kontinuit√§tsgleichung:**

$\begin{equation}
\frac{\partial{v_i}}{\partial{x_i}} = 0.
\end{equation}$

## Skalierung

### Skalierungskonstanten

Es gibt verschiedene M√∂glichkeiten, die Gleichungen zu skalieren. Entscheidend ist dabei die Wahl geeigneter Skalierungskonstanten. Wir definieren folgende Konstanten:

$$
\begin{aligned}
h_{sc} &= h, \\
t_{sc} &= \frac{h^2}{\kappa}, \\
v_{sc} &= \frac{\kappa}{h}, \\
\tau_{sc} &= \frac{\eta_0 \kappa}{h^2}, \\
T_{sc} &= \Delta T, \\
Q_{sc} &= \frac{\Delta T\, \kappa\, \rho_0\, c_p}{h^2},
\end{aligned}
$$

wobei:
- $h$ die M√§chtigkeit des Modells [m],
- $\kappa = \frac{k}{\rho_0 c_p}$ die thermische Diffusivit√§t [m¬≤/s],
- $\eta_0$ die Referenzviskosit√§t [Pa¬∑s],
- $\Delta T$ die Temperaturdifferenz zwischen Ober- und Unterseite des Modells [K] ist.

### Skalierungsgesetze

Mit den oben definierten Konstanten ergeben sich folgende Umrechnungen in dimensionslose Gr√∂√üen (Apostroph kennzeichnet skalierte Gr√∂√üen):

$$
\begin{aligned}
h &= h' \cdot h_{sc}, \\
t &= t' \cdot t_{sc}, \\
v &= v' \cdot v_{sc}, \\
\tau &= \tau' \cdot \tau_{sc}, \\
T &= T' \cdot T_{sc}, \\
Q &= Q' \cdot Q_{sc}.
\end{aligned}
$$

Wendet man diese Skalierung auf die Gleichungen (1)‚Äì(3) an, k√∂nnen viele physikalische Konstanten eliminiert werden. Es verbleiben dimensionslose Gleichungen, in denen zentrale Parameter als sogenannte **Schl√ºsselgr√∂√üen** zusammengefasst sind.

> **Hinweis:** Die folgende Vereinfachung gilt nur unter den hier angenommenen N√§herungen.

### Rayleigh-Zahl

Der wichtigste verbleibende Parameter ist die **Rayleigh-Zahl** $Ra$, definiert durch:

$\begin{equation}
Ra = \frac{\rho_0 g \alpha \Delta{T} h^3}{\eta_0 \kappa}.
\end{equation}$

Die Rayleigh-Zahl beschreibt das Verh√§ltnis der konvektionsf√∂rdernden Kr√§fte (thermischer Auftrieb) zu den konvektionshemmenden Kr√§ften (Viskosit√§t, thermische Diffusion). √úberschreitet $Ra$ einen kritischen Wert, wird eine thermisch geschichtete Lage instabil, und es setzt Konvektion ein.

Unabh√§ngig von den absoluten Werten der Parameter beschreibt eine bestimmte Rayleigh-Zahl stets eine klar definierte Konvektionsform. Daher ist es m√∂glich, mit skalierten numerischen oder experimentellen Modellen Vorg√§nge wie die Mantelkonvektion der Erde zu simulieren.

Das Skript ist so aufgebaut, dass entweder die Rayleigh-Zahl direkt vorgegeben wird (dann wird $\eta_0$ automatisch angepasst), oder die Referenzparameter variiert werden k√∂nnen, um gezielt eine bestimmte Rayleigh-Zahl zu erreichen.

F√ºr eine Rayleigh-Zahl von $10^5$ verwenden wir folgende Referenzwerte:

$$
\begin{aligned}
g &= 9{,}81\ \mathrm{m/s^2}, \\
\rho_0 &= 3300.0\ \mathrm{kg/m^3}, \\
k &= 4.125\ \mathrm{W/m/K}, \\
c_p &= 1250.0\ \mathrm{J/kg/K}, \\
\alpha &= 2{,}0 \times 10^{-5}\ \mathrm{K^{-1}}, \\
Q_0 &= 0.0\ \mathrm{W/m^3}, \\
\eta_0 &= 3{,}947725485 \times 10^{23}\ \mathrm{Pa{\cdot}s}, \\
\kappa &= 10^{-6}\ \mathrm{m^2/s}, \\
\Delta T &= 2500.0\ \mathrm{K}.
\end{aligned}
$$

## Skalierte Gleichungen

Die resultierenden dimensionslosen Gleichungen lauten:

### Impulserhaltung

$x$-Komponente:

$\begin{equation}
-\frac{\partial{P'}}{\partial{x'}}+\frac{\partial{\tau_{xj}'}}{\partial{x_j'}} = 0,
\end{equation}$

bzw. in Geschwindigkeitsform:

$\begin{equation}
-\frac{\partial{P'}}{\partial{x'}}+2\frac{\partial}{\partial{x'}}\eta'\frac{\partial{v_x'}}{\partial{x'}}+\frac{\partial}{\partial{y'}}\eta'\left(\frac{\partial{v_x'}}{\partial{y'}}+\frac{\partial{v_y'}}{\partial{x'}}\right) = 0.
\end{equation}$

$y$-Komponente:

$\begin{equation}
-\frac{\partial{P'}}{\partial{y'}}+\frac{\partial{\tau_{yj}'}}{\partial{x_j'}} = Ra T',
\end{equation}$

bzw. in Geschwindigkeitsform:

$\begin{equation}
-\frac{\partial{P'}}{\partial{y'}}+2\frac{\partial}{\partial{y'}}\eta'\frac{\partial{v_{y}'}}{\partial{y'}} +\frac{\partial}{\partial{x'}}\eta'\left(\frac{\partial{v_{y}'}}{\partial{x'}} + \frac{\partial{v_{x}'}}{\partial{y'}}\right) = Ra T'.
\end{equation}$

### Temperaturerhaltung

$\begin{equation}
\left(\frac{\partial{T'}}{\partial{t'}} + v_j' \frac{\partial{T'}}{\partial{x_j'}}\right) = \frac{\partial^2{T'}}{\partial{x^{'2}_i}} + Q'.
\end{equation}$

### Kontinuit√§tsgleichung

$\begin{equation}
\frac{\partial{v'_x}}{\partial{x}} + \frac{\partial{v'_y}}{\partial{y'}} = 0.
\end{equation}$

Im Wesentlichen √§ndert sich in der Impulserhaltungsgleichung nur die rechte Seite: Der Auftriebsterm wird durch das Produkt aus Rayleigh-Zahl und dimensionsloser Temperatur ersetzt.


## Die L√∂sung

### Aufgabenstellung

Vervollst√§ndige die Skalierung der Gleichungen durch:

- die Definition der **Skalierungskonstanten** sowie  
- die Anwendung der **Skalierungsgesetze**.

>**Beachte:** Neben der Anwendung der Skalierungsgesetze muss **auch die rechte Seite** der Impulserhaltung angepasst werden (siehe Gleichung $(8)$).  

F√ºr den Aufruf der Funktionen zur L√∂sung der Impulserhaltungs- und Diffusionsgleichung nehmen wir in der skalierten Form an, dass die folgenden **Referenzparameter** gleich **1** sind:

- $\eta_0 = 1$  
- $g = 1$  
- $\kappa = 1$  
- $c_p = 1$

Nach erfolgreicher Skalierung betrachten wir das Verhalten f√ºr **drei verschiedene Rayleigh-Zahlen**:

$$
Ra = 10^4,\quad 10^5,\quad \text{und} \quad 10^6.
$$

**Beachte:** Mit zunehmender Rayleigh-Zahl:

- steigen die **Str√∂mungsgeschwindigkeiten**,  
- die **Konvektion wird intensiver**,  
- die Strukturen wie *Slabs* und *Plumes* werden **feinr√§umiger**.

Daher muss die **Gitteraufl√∂sung** entsprechend angepasst werden, um **numerische Stabilit√§t** und **Genauigkeit** zu gew√§hrleisten.

> Eine gr√∂√üere Aufl√∂sung erh√∂ht jedoch den Rechenaufwand erheblich!

Die hier vorgegebene Aufl√∂sung ist f√ºr die genannten Rayleigh-Zahlen ausreichend.  
Allerdings lassen sich bei bestimmten numerischen Methoden bereits **erste Ungenauigkeiten** erkennen, weshalb in der Praxis oft eine **h√∂here Aufl√∂sung** empfehlenswert w√§re.

Im Unterschied zu den bisherigen √úbungen verwenden wir hier sogenannte **ver√§nderbare Strukturen** (*mutable structures*) in `Julia`.

Nun k√∂nnen wir die zeitlichen Parameter bestimmen. Dabei m√ºssen vor allem die maximale Zeitschrittl√§nge und die maximale Zeit bestimmt werden. 

Bei der maximalen Zeitschrittl√§nge, m√ºssen wir sicher gehen, dass die Stabilit√§tskriterien erf√ºllt sind (eigentlich nur wenn bestimmte finite Differenzenverfahren verwendet werden). Die Stabilit√§tkriterien lassen sich auch mit Hilfe von bestimmten Multiplikationsfaktoren $\Delta facc$ und $\Delta facd$ anpassen. 

Das hei√üt, wir m√ºssen die Zeitschrittl√§nge f√ºr das Diffusionsstabilit√§tskriterium und f√ºr das Courantkriterium bestimmen. Die maximale Zeitschrittl√§nge wird dann bestimmt durch das Minimum der beiden. 

Die maximale Zeit und die maximale Anzahl der Iterationen geben wir vorerst adhoc an. Im weiteren, sollte die Modellierung aufh√∂ren, wenn sich die Durchschnittsgeschwindigkeit statistisch gesehen nicht mehr ver√§ndert. 

````julia
# Zeitparameter ===================================================== #
T   =   TimeParameter(
    tmax    =   1000000.0,          #   [ Ma ]
    Œîfacc   =   0.9,                #   Courant time factor
    Œîfacd   =   0.9,                #   Diffusion time factor
    itmax   =   8000,               #   Maximum iterations; 30000
)
T.tmax      =   T.tmax*1e6*T.year    #   [ s ]
T.Œîc        =   T.Œîfacc * minimum((Œî.x,Œî.y)) / 
                    (sqrt(maximum(abs.(D.vx))^2 + maximum(abs.(D.vy))^2))
T.Œîd        =   T.Œîfacd * (1.0 / (2.0 * P.Œ∫ *(1.0/Œî.x^2 + 1/Œî.y^2)))

T.Œî         =   minimum([T.Œîd,T.Œîc])

Time        =   zeros(T.itmax)
Nus         =   zeros(T.itmax)
meanV       =   zeros(T.itmax)
meanT       =   zeros(T.itmax,NC.y+2)
find        =   0
# ------------------------------------------------------------------- #
````

````julia
# Koordinaten ======================================================= #
x       =   (
    c   =   LinRange(M.xmin+Œî.x/2,M.xmax-Œî.x/2,NC.x),
    ce  =   LinRange(M.xmin - Œî.x/2.0, M.xmax + Œî.x/2.0, NC.x+2),
    v   =   LinRange(M.xmin,M.xmax,NV.x),
)
y       =   (
    c   =   LinRange(M.ymin+Œî.y/2,M.ymax-Œî.y/2,NC.y),
    ce  =   LinRange(M.ymin - Œî.x/2.0, M.ymax + Œî.x/2.0, NC.y+2),
    v   =   LinRange(M.ymin,M.ymax,NV.y),
)
x1      =   (
    c2d     =   x.c .+ 0*y.c',
    v2d     =   x.v .+ 0*y.v', 
    vx2d    =   x.v .+ 0*y.ce',
    vy2d    =   x.ce .+ 0*y.v',
)
x   =   merge(x,x1)
y1      =   (
    c2d     =   0*x.c .+ y.c',
    v2d     =   0*x.v .+ y.v',
    vx2d    =   0*x.v .+ y.ce',
    vy2d    =   0*x.ce .+ y.v',
)
y   =   merge(y,y1)
# ------------------------------------------------------------------- #
````

Als Anfangsbedingungen nehmen wir ein mit der Tiefe, linear ansteigendes Temperaturprofil mit einer kleinen Gaussian Anomalie. Der Temperaturgradient mit der Tiefe, wird bestimmt durch die Temperaturdifferenz $\Delta{T}$ und die M√§chtigkeit unserer Schicht $h$. 

F√ºr den Fall, dass **nicht** Marker zur Advektion der Temperatur verwendet werden, wir die Anfangsbedingung mit der Routine ```IniTemperature!()``` berechnet.

````julia
# Anfangsbedingungen ================================================ #
# Temperatur ------
if FD.Method.Adv==:tracers 
    # Tracer Initialization ---
    # Need to implement incremental marker update first! 
    nmx,nmy     =   3,3
    noise       =   0
    nmark       =   nmx*nmy*NC.x*NC.y
    Aparam      =   :thermal
    MPC         =   (
        c       =   zeros(Float64,(NC.x,NC.y)),
        v       =   zeros(Float64,(NV.x,NV.y)),
        th      =   zeros(Float64,(nthreads(),NC.x,NC.y)),
        thv     =   zeros(Float64,(nthreads(),NV.x,NV.y)),
    )
    MPC1        = (
        PG_th   =   [similar(D.œÅ) for _ = 1:nthreads()],    # per thread
        PV_th   =   [similar(D.Œ∑v) for _ = 1:nthreads()],   # per thread
        wt_th   =   [similar(D.wt) for _ = 1:nthreads()],   # per thread
        wtv_th  =   [similar(D.wtv) for _ = 1:nthreads()],  # per thread
    )
    MPC     =   merge(MPC,MPC1)
    Ma      =   IniTracer2D(Aparam,nmx,nmy,Œî,M,NC,noise,Ini.p,phase)
    # RK4 weights ---
    rkw     =   1.0/6.0*[1.0 2.0 2.0 1.0]   # for averaging
    rkv     =   1.0/2.0*[1.0 1.0 2.0 2.0]   # for time stepping
    # Count marker per cell ---
    CountMPC(Ma,nmark,MPC,M,x,y,Œî,NC,NV,1)
    # Interpolate from markers to cell ---
    Markers2Cells(Ma,nmark,MPC.PG_th,D.œÅ,MPC.wt_th,D.wt,x,y,Œî,Aparam,œÅ)
else
    IniTemperature!(Ini.T,M,NC,D,x,y;Tb=P.Tbot,Ta=P.Ttop)
    if FD.Method.Adv==:slf
        D.T_exo    .=  D.T_ex
    end
end
# ------------------------------------------------------------------- #
````

Bei den Randbedingungen m√ºssen wir f√ºr die Temperatur- und die Impulsgleichung die Bedingungen festlegen. Dabei nehmen wir f√ºr die thermischen Randbedingungen an:

2. S√ºden   -   Dirichlet
3. Westen  -   Neumann
4. Osten   -   Neumann 

F√ºr die Geschwindigkeitsrandbedingungen nehmen wir f√ºr alle R√§nder *free slip* Bedingungen an. 

````julia
# Randbedingungen =================================================== #
# Temperatur ------
TBC     = (
    type    = (W=:Neumann, E=:Neumann,N=:Dirichlet,S=:Dirichlet),
    val     = (W=zeros(NC.y),E=zeros(NC.y),
                    N=P.Ttop.*ones(NC.x),S=P.Tbot.*ones(NC.x)))
# Geschwindigkeit ------
VBC     =   (
    type    =   (E=:freeslip,W=:freeslip,S=:freeslip,N=:freeslip),
    val     =   (E=zeros(NV.y),W=zeros(NV.y),S=zeros(NV.x),N=zeros(NV.x),
                vxE=zeros(NC.y),vxW=zeros(NC.y),vyS=zeros(NC.x),vyN=zeros(NC.x)),
)
# ------------------------------------------------------------------- # 
````

Nun wird die *Rayleigh* Zahl oder die Referenzviskosit√§t berechnet: 

````julia
# Rayleigh Zahl Bedingungen ========================================= #
if P.Ra < 0
    # Falls die Rayleigh Zahl nicht explizit angegeben wird, dann 
    # wird sie hier berechnet
    P.Ra     =   P.œÅ‚ÇÄ*P.g*P.Œ±*P.ŒîT*S.hsc^3/P.Œ∑‚ÇÄ/P.Œ∫
else
    # Falls die Rayleigh Zahl explizit angegeben ist, dann wird hier 
    # die Referenzviskositaet Œ∑‚ÇÄ angepasst. 
    P.Œ∑‚ÇÄ     =   P.œÅ‚ÇÄ*P.g*P.Œ±*P.ŒîT*S.hsc^3/P.Ra/P.Œ∫
end
# =================================================================== #
filename    =   string("12_ThermalConvection_",P.Ra[1],
                        "_",NC.x,"_",NC.y,
                        "_",Ini.T,"_",FD.Method.Adv,"_",FD.Method.Diff,
                        "_",FD.Method.Mom)
````

Im folgenden werden die Parameter f√ºr die linearen Gleichungssysteme bestimmt (abh√§ngig von dem gew√§hlten finite Differenzen Verfahren; f√ºr die Impulserhaltung verwenden wir hier nur die direkte L√∂sungsmethode): 

````julia
# Lineares Gleichungssystem ========================================= #
# Impulserhaltung (IEG) ------
off    = [  NV.x*NC.y,                          # vx
            NV.x*NC.y + NC.x*NV.y,              # vy
            NV.x*NC.y + NC.x*NV.y + NC.x*NC.y ] # Pt

Num    =    (
    Vx  =   reshape(1:NV.x*NC.y, NV.x, NC.y), 
    Vy  =   reshape(off[1]+1:off[1]+NC.x*NV.y, NC.x, NV.y), 
    Pt  =   reshape(off[2]+1:off[2]+NC.x*NC.y,NC...),
    T   =   reshape(1:NC.x*NC.y, NC.x, NC.y),
)
ndof    =   maximum(Num.T)
if FD.Method.Mom==:dc
    niterM  =   50
    œµM      =   1e-10
    KM      =   ExtendableSparseMatrix(ndof,ndof)
end        
# Energieerhaltung (EEG) ------
if FD.Method.Diff==:implicit || FD.Method.Diff==:CNA
    if FD.Method.Diff==:CNA
        K1      =   ExtendableSparseMatrix(ndof,ndof)
        K2      =   ExtendableSparseMatrix(ndof,ndof)
    else
        K       =   ExtendableSparseMatrix(ndof,ndof)
    end
    rhs         =   zeros(ndof)
elseif FD.Method.Diff==:dc
    niter       =   10
    œµ           =   1e-10
    k           =   (x=ones(NC.x+1,NC.y), y=ones(NC.x,NC.y+1))
    K           =   ExtendableSparseMatrix(ndof,ndof)
    R           =   zeros(Float64,NC...)
    ‚àÇT          =   (‚àÇx=zeros(NC.x+1, NC.y), ‚àÇy=zeros(NC.x, NC.y+1))
    q           =   (x=zeros(NC.x+1, NC.y), y=zeros(NC.x, NC.y+1))
end
# ------------------------------------------------------------------- #
````

Nun k√∂nnen wir die Gleichungen in der Zeitschleife l√∂sen. Dabei m√ºssen die folgenden Schritte ber√ºcksichtig werden: 

1. Berechnung der Zeit
2. Initialisierung des unbekannten Vektors und der rechten Seite f√ºr die Impulserhaltung
3. Null Setzung der Geschwindigkeits- und Druckfelder
4. Aufstellung der Koeffizientenmatrix mit Hilfe von ```Assemblyc```
5. Aktualisierung der rechten Seite mit Hilfe von ```updaterhs```
6. L√∂sen der Impulserhaltung
7. Update der Geschwindigkeits- und Druckfelder
8. Bestimmung der Geschwindikeiten auf den *Centroids* 
9. Anpassung der maximalen Zeitschrittl√§nge
10. Darstellung der Geschwindigkeit, Temperatur, und Dichte
11. L√∂sen der Advektionsgleichung
12. L√∂sen der Diffusionsgleichung
13. Berechnung der statistischen Analysewerte (*Nusselt* Zahl, $V_{RMS}$, etc.)
14. Aktualisierung der Dichte mit der *Zustandsgleichung*

````julia
# Zeitschleife ====================================================== #
for it = 1:T.itmax
    œá       =   zeros(maximum(Num.Pt))      #   Unbekannten Vektor IEG
    rhsM    =   zeros(maximum(Num.Pt))      #   Rechte Seite IEG
    if it>1
        Time[it]  =   Time[it-1] + T.Œî
    end
    @printf("Time step: #%04d, Time [non-dim]: %04e\n ",it,
                    Time[it])
    # IEG ------
    D.vx    .=  0.0
    D.vy    .=  0.0 
    D.Pt    .=  0.0
    if FD.Method.Mom==:direct
        # Update K ---
        KM      =   Assemblyc(NC, NV, Œî, 1.0, VBC, Num)
        # Update RHS ---
        # rechte Seite definiert durch die Boussinesq Ann√§herung
        rhsM    =   updaterhsc( NC, NV, Œî, 1.0, -P.Ra*D.T, -1.0, VBC, Num )
        # L√∂sen des linearen Gleichungssystems ---
        œá       =   KM \ rhsM
        # Update unbekannte Variablen ---
        D.vx[:,2:end-1]     .=  œá[Num.Vx]
        D.vy[2:end-1,:]     .=  œá[Num.Vy]
        D.Pt                .=  œá[Num.Pt]
    elseif FD.Method.Mom==:dc
        @. D.œÅ  =   -P.Ra*D.T
        # Anfangsresiduum ------
        for iter = 1:niterM
            Residuals2Dc!(D,VBC,Œµ,œÑ,divV,Œî,1.0,1.0,Fm,FPt)
            rhsM[Num.Vx]    =   Fm.x[:]
            rhsM[Num.Vy]    =   Fm.y[:]
            rhsM[Num.Pt]    =   FPt[:]
            @printf("||R_M|| = %1.4e\n", norm(rhsM)/length(rhsM))
            norm(rhsM)/length(rhsM) < œµM ? break : nothing
            # Update K ------
            KM      =   Assemblyc(NC, NV, Œî, 1.0, VBC, Num)
            # L√∂sen des lineare Gleichungssystems ------
            œá      =   - KM \ rhsM
            # Update unbekante Variablen ------
            D.vx[:,2:end-1]     .+=  œá[Num.Vx]
            D.vy[2:end-1,:]     .+=  œá[Num.Vy]
            D.Pt                .+=  œá[Num.Pt]
        end
        D.œÅ  =   ones(NC...)
    end
    # ======
    # Berechnung der Geschwindikeit auf den Centroids ------
    for i = 1:NC.x
        for j = 1:NC.y
            D.vxc[i,j]  = (D.vx[i,j+1] + D.vx[i+1,j+1])/2
            D.vyc[i,j]  = (D.vy[i+1,j] + D.vy[i+1,j+1])/2
        end
    end
    @. D.vc        = sqrt(D.vxc^2 + D.vyc^2)
    # ---
    @show(maximum(D.vc))
    @show(minimum(D.Pt))
    @show(maximum(D.Pt))
    # Berechnung der Zeitschrittl√§nge =============================== #
    T.Œîc        =   T.Œîfacc * minimum((Œî.x,Œî.y)) / 
            (sqrt(maximum(abs.(D.vx))^2 + maximum(abs.(D.vy))^2))
    T.Œîd        =   T.Œîfacd * (1.0 / (2.0 *(1.0/Œî.x^2 + 1/Œî.y^2)))
    T.Œî         =   minimum([T.Œîd,T.Œîc])
    if Time[it] > T.tmax
        T.Œî         =   T.tmax - Time[it-1]
        Time[it]    =   Time[it-1] + T.Œî
        it          =   T.itmax
    end
    # Plot ========================================================== #
    if mod(it,10) == 0 || it == T.itmax || it == 1
        p = heatmap(x.c,y.c,D.T',
                xlabel="x",ylabel="y",colorbar=true,
                title="Temperature",color=cgrad(:lajolla),
                aspect_ratio=:equal,xlims=(M.xmin, M.xmax),
                ylims=(M.ymin, M.ymax),
                layout=(2,1),subplot=1)
        heatmap!(p,x.c,y.c,D.vc',color=:imola,
            xlabel="x[km]",ylabel="y[km]",colorbar=true,
            title="Velocity",aspect_ratio=:equal,
            xlims=(M.xmin, M.xmax), 
            ylims=(M.ymin, M.ymax),
            layout=(2,1),subplot=2)
        quiver!(p,x.c2d[1:Pl.qinc:end,1:Pl.qinc:end],
            y.c2d[1:Pl.qinc:end,1:Pl.qinc:end],
            quiver=(D.vxc[1:Pl.qinc:end,1:Pl.qinc:end].*Pl.qsc,
                    D.vyc[1:Pl.qinc:end,1:Pl.qinc:end].*Pl.qsc),
            la=0.5,color="black",
            layout=(2,1),subplot=2)
        if save_fig == 1
            Plots.frame(anim)
        elseif save_fig == 0
            display(p)
        end
    end
    # --------------------------------------------------------------- #
    # Advektion ===================================================== #
    if FD.Method.Adv==:upwind
        upwindc2D!(D.T,D.T_ex,D.vxc,D.vyc,NC,T.Œî,Œî.x,Œî.y)            
    elseif FD.Method.Adv==:slf
        slfc2D!(D.T,D.T_ex,D.T_exo,D.vxc,D.vyc,NC,T.Œî,Œî.x,Œî.y)
    elseif FD.Method.Adv==:semilag
        semilagc2D!(D.T,D.T_ex,D.vxc,D.vyc,[],[],x,y,T.Œî)
    elseif FD.Method.Adv==:tracers
        # Advect tracers ---
        @printf("Running on %d thread(s)\n", nthreads())  
        AdvectTracer2D(Ma,nmark,D,x,y,T.Œî,Œî,NC,rkw,rkv,1)
        CountMPC(Ma,nmark,MPC,M,x,y,Œî,NC,NV,it)
        # Interpolate phase from tracers to grid ---
        Markers2Cells(Ma,nmark,MPC.PG_th,D.œÅ,MPC.wt_th,D.wt,x,y,Œî,Aparam,œÅ)
    end
    # --------------------------------------------------------------- #
    # Diffusion ===================================================== #
    if FD.Method.Diff==:explicit
        ForwardEuler2Dc!(D, 1.0, Œî.x, Œî.y, T.Œî, D.œÅ, 1.0, NC, TBC)
    elseif FD.Method.Diff==:implicit
        BackwardEuler2Dc!(D, 1.0, Œî.x, Œî.y, T.Œî, D.œÅ, 1.0, NC, TBC, rhs, K, Num)
    elseif FD.Method.Diff==:CNA
        CNA2Dc!(D, 1.0, Œî.x, Œî.y, T.Œî, D.œÅ, 1.0, NC, TBC, rhs, K1, K2, Num)
    elseif FD.Method.Diff==:ADI
        ADI2Dc!(D, 1.0, Œî.x, Œî.y, T.Œî, D.œÅ, 1.0, NC, TBC)
    elseif FD.Method.Diff==:dc
        D.T0    .=  D.T
        for iter = 1:niter
            # Evaluate residual
            ComputeResiduals2D!(R, D.T, D.T_ex, D.T0, ‚àÇT, q, D.œÅ, D.cp, k, TBC, Œî, T.Œî)
            @printf("||R|| = %1.4e\n", norm(R)/length(R))
            norm(R)/length(R) < œµ ? break : nothing
            # Assemble linear system
            K  = AssembleMatrix2D(D.œÅ, D.cp, k, TBC, Num, NC, Œî, T.Œî)
            # Solve for temperature correction: Cholesky factorisation
            Kc = cholesky(K.cscmatrix)
            # Solve for temperature correction: Back substitutions
            Œ¥T = -(Kc\R[:])
            # Update temperature
            @. D.T += Œ¥T[Num.T]
        end        
    end
    # --------------------------------------------------------------- #
    # W√§rmefluss an der Oberfl√§che ================================== #
    @. D.ŒîTbot  =   
         (((D.T_ex[2:end-1,2]+D.T_ex[2:end-1,3])/2.0) - 
        ((D.T_ex[2:end-1,2]+D.T_ex[2:end-1,1])/2.0)) / Œî.y
    @. D.ŒîTtop  =   
        (((D.T_ex[2:end-1,end-2]+D.T_ex[2:end-1,end-1]) / 2.0) - 
        ((D.T_ex[2:end-1,end-1]+D.T_ex[2:end-1,end]) / 2.0)) / Œî.y
    Nus[it]     =   mean(D.ŒîTtop)
    meanT[it,:] =   mean(D.T_ex,dims=1)
    meanV[it]   =   mean(D.vc)
    # --------------------------------------------------------------- #
    # Check break =================================================== #
    # If the maximum time is reached or if the models reaches steady 
    # state the time loop is stoped! 
    if Time[it] > 0.0038
        epsC    =   1e-3; 
        ind     =   findfirst(Time .> 
                        (Time[it] - 0.0038))
        epsV    =   std(meanV[ind:it])
        if save_fig == 1
            plot!(ks,(it,log10((epsV))),
                xlabel="it",ylabel="log‚ÇÅ‚ÇÄ(Œµ·µ•)",label="",
                markershape=:circle,markercolor=:black)
        end
        find    =   it
        @printf("Œµ_V = %g, Œµ_C = %g \n",epsV,epsC)
        if Time[it] >= T.tmax
            @printf("Maximum time reached!\n")
            find    =   it
            break
        elseif (epsV <= epsC)
            @printf("Convection reaches steady state!\n")
            find    =   it
            break
        end
    end
    # --------------------------------------------------------------- #
    @printf("\n")
end
````
